```python
# create a class to collect questions and answers from the data list
class Question:
    def __init__(self, text, answer):
        self.text = text
        self.answer = answer
```

```python
# create a list of question and answer class
questions = []
for element in question_data:
    question = element['question'] # get question
    answer = element['correct_answer'] # get answer
    new_question = Question(question, answer) # create a data object
    questions.append(new_question) # append it to the questions data list
```

```python
class QuizBrain:
    '''
    Brain of the game to under the followings:
    - check if any more question available
    - get new question and check answer
    - check answer and outout the results
    '''
    def __init__(self, questions):
        self.questions = questions
        self.score = 0
        self.counter = 0

    def has_question(self):
        '''
        checks if there are any more question in the data list
        '''
        return self.counter < len(self.questions)

    def get_new_question(self):
        '''
        get new question from the data bank and check answer
        '''
        next_question = self.questions[self.counter].text
        correct_answer = self.questions[self.counter].answer
        self.counter += 1
        user_answer = input(f'Q.{self.counter}: {next_question}: Answer - True or False: ')
        self.check_answer(user_answer, correct_answer)
        
    def check_answer(self, user_answer, correct_answer):
        '''
        check user answer
        update the score 
        output the result
        '''
        if user_answer.lower() == correct_answer.lower():
            print('Correct Answer!')
            self.score += 1 # increase the score
        else:
            print('Wrong Answer')

        print(f"The correct answer was: {correct_answer}.")
        print(f"Your current score is: {self.score}/{self.counter}")
        print("\n")
```

```python
# create a quiz object 
quiz = QuizBrain(questions)

# start to loop to play the game
while quiz.has_question():
    quiz.get_new_question()

# output the final results
print("You've completed the quiz")
print(f"Your final score was: {quiz.score}/{quiz.counter}")
```
