```python
import random
from IPython.display import clear_output
```

```python
class Board:
    def __init__(self):
        self.board = [[' ', ' ', ' '],
                      [' ', ' ', ' '],
                      [' ', ' ', ' ']]
        

    def _display_board(self):
        print(' ' + self.board[0][0]  + ' | ' + self.board[0][1] + ' | ' + self.board[0][2])
        print('-----------')
        print(' ' + self.board[1][0]  + ' | ' + self.board[1][1] + ' | ' + self.board[1][2])
        print('-----------')
        print(' ' + self.board[2][0]  + ' | ' + self.board[2][1] + ' | ' + self.board[2][2])
```

```python
class Play_Game(Board):
    def __init__(self):
        super().__init__()
        self.win = False
        self.player = None

    def _toss(self):
        num = random.choice([0, 1])

        if num == 0:
            self.player = 'player_1'
        else:
            self.player = 'player_2'
        
        return self.player

    def _game_outcome(self):

        board_col = []
        board_dia = [[self.board[0][0], self.board[1][1], self.board[2][2]],
                     [self.board[0][-1], self.board[1][-2], self.board[2][-3]]]

        for i in range(3):
            col = []
            for j in range(3):
                col.append(self.board[j][i])
            board_col.append(col)

        for i in range(3):
            row = set(self.board[i])
            col = set(board_col[i])
            if (len(row) == 1) and (' ' not in row):
                self.win = True
                return self.win
            elif (len(col) == 1) and (' ' not in col):
                self.win = True
                return self.win

        for dia in board_dia:
            dia = set(dia)
            if (len(dia) == 1) and (' ' not in dia):
                self.win = True
                return self.win

        return False

    def play(self):

        self.player = self._toss()
        game_over = False
        self._display_board()

        while game_over == False:

            while True:
                
                print('It is {} Turn'.format(self.player))
                entry_row = input('Enter row number between 1 and 3: ')
                entry_col = input('Enter col number betweem 1 and 3: ') 

                try:
                    entry_row = int(entry_row) - 1
                    entry_col = int(entry_col) - 1

                    if 0 <= entry_row <= 2 and 0 <= entry_col <= 2:
                        break
                    else:
                        print('Please enter between 1 and 3')
                except:
                    print('Please enter between 1 and 3')


            if self.board[entry_row][entry_col] == ' ':
                if self.player =='player_1':
                    self.board[entry_row][entry_col] = 'X'
                    game_over = self._game_outcome()
                    if not game_over:
                        self.player ='player_2'

                else:
                    self.board[entry_row][entry_col] = 'O'
                    game_over = self._game_outcome()
                    if not game_over:
                        self.player ='player_1'
            else:
                print('The place has already been taken. Please choose another one')

            clear_output(wait=True)
            self._display_board()

            if game_over:
                print('\nHurreyyyyyyy! The Winner is {}'.format(self.player))

            elif not any(' ' in sublist for sublist in self.board):
                game_over = True
                print('\nGame is a Draw')

        
```
