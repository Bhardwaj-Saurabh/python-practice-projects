```python
def select_celebrity(data):
    
    celebrity = random.choice(data)
    return celebrity
```

```python
score = 0

game_over = False
while not game_over:
    celebrity_1 = select_celebrity(data)
    celebrity_2 = select_celebrity(data)

    while celebrity_1 == celebrity_2:
        celebrity_2 = select_celebrity(data)

    print('Compare A: {}, a {} from {}'.format(celebrity_1['name'], celebrity_1['description'], celebrity_1['country']))
    print(vs)
    print('Against B: {}, a {} from {}'.format(celebrity_2['name'], celebrity_2['description'], celebrity_2['country']))

    celebrity_1_followers = celebrity_1['follower_count']
    celebrity_2_followers = celebrity_2['follower_count']

    user_input = input('\nWho has more followers: Type A or B: ')

    if celebrity_1_followers > celebrity_2_followers and user_input == 'A':
        score += 1
    elif celebrity_1_followers < celebrity_2_followers and user_input == 'B':
        score += 1
    else:
        game_over = True
    clear_output()
    print(f'Your Score is {score}')
    
```
