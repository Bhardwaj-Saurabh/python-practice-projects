```python
# Alphabet list
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', \
            'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
```

```python
# define the caesar function to encode or decode the message
def caesar(text, shift, direction):
    '''
    This function return a encoded or decoded text based on the direction 
    given by the user.

    *args*
    text: text to encode or decode
    shift: how much shift user want to have
    direction: encode or decode

    *return*
    encoded or decoded text
    '''

    returned_text = ''
    for letter in text:
        if letter in alphabet:
            idx = alphabet.index(letter)
            if direction == 'encode':
                if (idx+shift) < 26:
                    new_letter = alphabet[idx + shift]
                else:
                    new_letter = alphabet[idx + shift - 26]
                returned_text += new_letter
            elif direction == 'decode':
                new_letter = alphabet[idx - shift]
                returned_text += new_letter
        else:
            returned_text += letter

    return returned_text
```

```python
program_end = False

while not program_end:
    #Take input from the user
    direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
    text = input("Type your message:\n").lower()
    shift = int(input("Type the shift number:\n"))

    # call the function 
    returned_text = caesar(text, shift, direction)

    # output the final text
    print(f'The {direction}d message is "{returned_text}"')

    proceed = input('\nDo you want to continue: ').lower()
    if proceed not in ['yes', 'y']:
        print('Good Bye')
        program_end = True
```
