```python
# import necessary libraries
import random
from IPython.display import clear_output
```

```python
logo ='''

  / _ \_   _  ___  ___ ___  /__   \ |__   ___    /\ \ \_   _ _ __ ___ | |__   ___ _ __ 
 / /_\/ | | |/ _ \/ __/ __|   / /\/ '_ \ / _ \  /  \/ / | | | '_ ` _ \| '_ \ / _ \ '__|
/ /_\\| |_| |  __/\__ \__ \  / /  | | | |  __/ / /\  /| |_| | | | | | | |_) |  __/ |   
\____/ \__,_|\___||___/___/  \/   |_| |_|\___| \_\ \/  \__,_|_| |_| |_|_.__/ \___|_|  

'''
```

```python
def difficulty_level():
    '''
    this function return the attempts avaiable to player based on the difficulty level
    '''
    level = input('Do you want to play in easy or hard mode: type easy or hard: \n').lower()
    if level == 'easy':
        return 10
    elif level == 'hard':
        return 5
```

```python
def play_game():
    '''
    this function return the final outcome of the same
    '''
    # get a random number to guess for player
    number_to_guess = random.choice(range(1, 101))
    # condition to check whether the game is over or not
    game_over = False
    # get the number of attempts available 
    attempts = difficulty_level()

    while not game_over:
        # ask user to guess an integer number
        user_guess = int(input('Please guess a integer number beteen 1 and 100: \n'))
        if user_guess < number_to_guess - int((number_to_guess/100) * 30):
            print("It is too low")
        elif user_guess < number_to_guess - int((number_to_guess/100) * 5):
            print("It is low")
        elif user_guess > number_to_guess + int((number_to_guess/100) * 30):
            print("It is too high")
        elif user_guess > number_to_guess + int((number_to_guess/100) * 5):
            print("It is high")
        elif user_guess == number_to_guess:
            print("You have guessed it correctly. You WON")
            game_over = True
        elif attempts == 0:
            print(f"You have ran out of the guess. You LOST.\nThe right answer is {number_to_guess}.")
            game_over = True
        else:
            print("You are close")
        
        # reduce the number of attempts remaining by 1
        attempts -= 1
    
```

```python
while input('\nDo you want to play: type y or n:\n').lower() == 'y':
    clear_output()
    print(logo)
    play_game()
    
```
