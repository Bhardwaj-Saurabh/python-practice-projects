```python
def highest_bid_checker(bids):
    winner = ''
    largest_bid = 0
    for name, bid in bids.items():
        if bid >= largest_bid:
            largest_bid = bid
            winner = name
    print(f'The winner is {winner.title()}.')

print('The item for the bid is\n', logo)
```

```python
# create empty dictionary to store the bidders information 
all_bids = {}
# define a boolean variable to check whether there is no more person to bid
bid_end = False

while not bid_end:
    
    # Ask user to enter their name
    name = input('Please enter you name: ')
    # Ask user to enter a bid amount
    amount = int(input('Please enter your bid amount: '))
    # add the entry to all_bid dictionary
    all_bids[name] = amount
    # Ask user to confirm if there are any more people to bid
    next_entry = input('Are there any more people to bid: Type yes or no.\n').lower()
    # clear output to hide information
    clear_output()
    # check if another person is there to bid
    if next_entry == 'no':
        bid_end = True
        highest_bid_checker(all_bids)
    elif next_entry == 'yes':
        # clear output to hide information
        clear_output()
    
```
